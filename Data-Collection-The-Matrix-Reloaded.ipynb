## Webscrapping

```python
#Import packages
import time
import requests
from bs4 import BeautifulSoup
import requests
from tqdm import tqdm
```

```python
#parse html file
source = requests.get('https://www.imsdb.com/scripts/Matrix-Reloaded,-The.html').text
soup = BeautifulSoup(source,'xml')
```

```python
print(soup.prettify())
```

```python
#Getting the entire movie script
movie_text = soup.find(class_ ='scrtext')
```

```python
#Checking
movie_text
```

```python
movie_script = list(movie_text)[1]
```

```python
movie_script
```

## Transforming data (readable format)

```python
#Saving the script as a list
test = []
for w in movie_script:
    test.append(w)
```

```python
test[5]
```

```python
#taking out the first useless lines
test = test[5:]
```

```python
test
```

```python
len(test)
```

```python
str(test[0]).startswith('<b>')
```

```python
#Create an empty list to store the pair speaker/words
script_list = []
#create a counter in order to go thru the list of scripts
indx = 0
#going thru the list w/out for loops using the index for comparison
#check if index is even
#if even and contain <b> - create a list with speaker and what he said
while indx < len(test):
    if str(test[indx]).startswith('<b>') and str(test[indx+1]).startswith('<b>')==False:
        script_list.append([test[indx],test[indx+1]])
        indx +=2
    elif str(test[indx]).startswith('<b>') and str(test[indx+1]).startswith('<b>'):
        script_list.append([test[indx],"Nothing"])
        indx +=1
```

```python
len(script_list)
```

```python
#creating a DataFrame with the scripts
import pandas as pd
df = pd.DataFrame(script_list,columns=['Speaker','Text'])
```

```python
df.head()
```

```python
df.tail()
```

## Cleaning

```python
def clean(line):
    line = line.replace('\t',' ')
    line = line.replace('\n','')
    line = line.replace('<b>','')
    line = line.replace('</b>','')
    return line
```

```python
df.Speaker = df.Speaker.apply(lambda x: clean(str(x)))
df.Text = df.Text.apply(lambda x: clean(str(x)))
```
